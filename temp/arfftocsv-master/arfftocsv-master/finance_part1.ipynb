```python
s_c = samsung.reset_index()['Close'].tolist()
s_c
```

```python
s_c_2d = []
s_c_after1 = []
s_c_after3 = []
s_c_after7 = []
i = 0

data_length = 70  # day length for one sliced data
data_count = int(3494/data_length)  # total 3494 daily stock datas and floor data amount by length
for _ in range(data_count):
    s_c_2d.append(s_c[data_length*i : data_length*(i+1)])
    s_c_after1.append(s_c[data_length*(i+1)])
    s_c_after3.append(s_c[data_length*(i+1)+2])
    s_c_after7.append(s_c[data_length*(i+1)+6])
    i += 1
```

```python
after1_percent = []
after3_percent = []
after7_percent = []
total_percent = []

for i in range(data_count):
    after1_percent.append((s_c_after1[i]-s_c_2d[i][-1]) / s_c_2d[i][-1] * 100)
    after3_percent.append((s_c_after3[i]-s_c_2d[i][-1]) / s_c_2d[i][-1] * 100)
    after7_percent.append((s_c_after7[i]-s_c_2d[i][-1]) / s_c_2d[i][-1] * 100)

total_percent.append(after1_percent)
total_percent.append(after3_percent)
total_percent.append(after7_percent)
```

```python
# total_T = np.array(total_percent).T.tolist()
# plt.plot(total_T)

answer = []
for i in range(len(total_percent)):
    tmp = []
    for per in total_percent[i]:
        if abs(per) < 3:
            tmp.append(0)
        elif per > 3:
            tmp.append(1)
        elif per < -3:
            tmp.append(-1)
    answer.append(tmp)
    
len(answer[2])
```

```python
# 전수저장
s_c_full = []
p = 0
for p in range(3494-data_length-6):
    s_c_full.append(s_c[0+p : data_length+p])
    p += 1

len(s_c_full)
```

```python
after1_full = []
after3_full = []
after7_full = []

# 여기 아래는 수정필요
for i in range(len(s_c_full)):
    after1_full.append((s_c[data_length+i]-s_c_full[i][-1]) / s_c_full[i][-1] * 100)
    after3_full.append((s_c[data_length+i+2]-s_c_full[i][-1]) / s_c_full[i][-1] * 100)
    after7_full.append((s_c[data_length+i+6]-s_c_full[i][-1]) / s_c_full[i][-1] * 100)

after7_full
```

```python
count_3 = pd.DataFrame(columns=[0,1,2], index=[0,1,2])
count_3.loc[:,:] = 0
# count.loc[1,0] += 1
total_full = []
total_full.append(after1_full)
total_full.append(after3_full)
total_full.append(after7_full)

for i in range(len(total_full)):
    for per in total_full[i]:
        if abs(per) < 3:
            count_3.loc[i,2] += 1
        elif per > 3:
            count_3.loc[i,0] += 1
        elif per < -3:
            count_3.loc[i,1] += 1
count_3
```

```python
count_2 = pd.DataFrame(columns=[0,1,2], index=[0,1,2])
count_2.loc[:,:] = 0
# count.loc[1,0] += 1
total_full = []
total_full.append(after1_full)
total_full.append(after3_full)
total_full.append(after7_full)

for i in range(len(total_full)):
    for per in total_full[i]:
        if abs(per) < 2:
            count_2.loc[i,2] += 1
        elif per > 2:
            count_2.loc[i,0] += 1
        elif per < -2:
            count_2.loc[i,1] += 1
count_2
```
