```python
kospi200_dict = {}
kospi200_list = []
kospi200_korean = []
Url = "https://finance.naver.com/sise/entryJongmok.nhn?&page="
for i in range(1,21):
    url = Url + str(i)
    res = req.urlopen(url).read()
    soup = BeautifulSoup(res, "html.parser")

    kospi_list = soup.select("td.ctg")
    for name in kospi_list:
        krx = name.a.get('href')[-6:]
        kospi200_dict[krx] = name.string
        kospi200_list.append(krx)
        kospi200_korean.append(name.string)
        # print('name : ', name.string)
        # print('KRX : ', name.a.get('href')[-6:])
```

```python
print(len(kospi200_list))
```

```python
train_data = []
tmp = []
stock_name = []
for data in kospi200_list:
    tmp = fdr.DataReader(data, '2014-10-01', '2018-09-30').reset_index()['Close'].tolist()
    if len(tmp) != 981:  # fdr 에서 우리금융지주 316140이 데이터가 없음..
        print(data)
        print(len(tmp))
        continue
    tmp = minmax_scale(tmp, axis=0)
    train_data.append(tmp)
    stock_name.append(kospi200_dict[data])
```

```python
# define the number of features
ncol = train_data.shape[1]
ncol
```

```python
### Define the encoder dimension
encoding_dim = 30
```

```python
input_dim = Input(shape = (ncol, ))

# Encoder Layers
encoded1 = Dense(750, activation = 'relu')(input_dim)
encoded2 = Dense(500, activation = 'relu')(encoded1)
encoded3 = Dense(250, activation = 'relu')(encoded2)
encoded4 = Dense(150, activation = 'relu')(encoded3)
encoded5 = Dense(75, activation = 'relu')(encoded4)
encoded6 = Dense(encoding_dim, activation = 'relu')(encoded5)

# Decoder Layers
decoded1 = Dense(75, activation = 'relu')(encoded6)
decoded2 = Dense(150, activation = 'relu')(decoded1)
decoded3 = Dense(250, activation = 'relu')(decoded2)
decoded4 = Dense(500, activation = 'relu')(decoded3)
decoded5 = Dense(750, activation = 'relu')(decoded4)
decoded6 = Dense(ncol, activation = 'sigmoid')(decoded5)

# Combine Encoder and Deocder layers
autoencoder = Model(inputs = input_dim, outputs = decoded6)

# Compile the Model
autoencoder.compile(optimizer = 'adadelta', loss = 'binary_crossentropy')
```

```python
autoencoder.summary()
```

```python
# 여기서 validation data가 의미하는건 뭐지?
autoencoder.fit(train_data, train_data, nb_epoch = 50, batch_size = 8, shuffle = False, validation_data = (train_data, train_data))
```
